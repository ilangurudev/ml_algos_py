# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np

from typing import Union
import collections
import math
import itertools

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
from plotnine import *
```

```python
df_train_org = pd.read_csv("data/train.csv")
df_test_org = pd.read_csv("data/test.csv")
df_train_org['is_affected'] = (df_train_org['area'] > 0)*1
```

```python
sns.pairplot(df_train_org)
```

The pair plot has too many variables and hence is unclear. We need boxplots that compare between the two groups. 

```python
df_train_org['is_affected'] = (df_train_org['area'] > 0)
df_train_mlt = df_train_org.melt(id_vars='is_affected')
# print(df_train_mlt.head())
(ggplot(aes(x='variable', 
            y='value', 
            fill='is_affected'), 
        data=df_train_mlt)
 + geom_boxplot() #aes(outlier_color='factor(is_affected)')
#  + coord_flip()
)
```

But since the values are not scaled, we are not really able to say. 

```python
m = df_train_org.mean()
sd = df_train_org.std()
df_train_cp = df_train_org.copy()
df_train_cp = (df_train_cp - m)/sd
df_train_cp['is_affected'] = df_train_org['is_affected'].copy()
df_train_mlt = df_train_cp.melt(id_vars='is_affected')
# print(df_train_mlt.head())
(ggplot(aes(x='variable', 
            y='value', 
            fill='is_affected',
#             color='is_affected'
           ), 
        data=df_train_mlt)
 + geom_boxplot(aes(outlier_size=0))
#  + geom_jitter(height=0)
#  + coord_flip() 
 + ylim(-5, 5) 
 + theme_light())    
```

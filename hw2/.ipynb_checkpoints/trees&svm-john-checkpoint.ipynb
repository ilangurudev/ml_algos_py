A function to scale the data by its mean and variance

```python
def data_scaling(train_x, test_x):
    mean = train_x.mean(0)
    std = train_x.std(0)
    train_x = (train_x - mean)/std
    test_x = (test_x - mean)/std
    return train_x, test_x
```

## Question 3(b.i) : A KNN class contains the k-cross-validator, fit, predict and accuracy of the model

```python
class KNN(object):
    
    def __init__(self, dist):
        self.dist = dist
        print ("KNN initialized...")
    
    def fit(self, train_x, train_y):
        self.train_x = train_x
        self.train_y = train_y
        
        # create a hashmap with indices
        knn_train_x = {}
        knn_train_y = {}
        
        for i in range(len(train_x)):
            knn_train_x[i] = train_x[i]
            knn_train_y[i] = train_y[i][0]
            
        self.knn_train_x = knn_train_x
        self.knn_train_y = knn_train_y
        
    
    def k_cross_validator(self, train_x, train_y, k_NN, fold = 5): 
        #concatenate train_x and train_y and shuffle them
        train_xy = np.concatenate((train_x, train_y), axis = 1)
        np.random.shuffle(train_xy)
        train_x = train_xy[:, [x for x in range(np.shape(train_xy)[1]-1)]]
        train_y = train_xy[:, 10].reshape(np.shape(train_y)[0], 1).astype(int)
        
        #get the range for example window = 466/5 = 93
        win = int(np.shape(train_y)[0]/fold)
        
        k_acc = []
        for j in range(len(k_NN)):
            val_acc = 0
            if k_NN[j] < 10:
                print ("Fit 0"+str(k_NN[j])+"-NN Fold: ", end =" ")
            else:
                print ("Fit "+str(k_NN[j])+"-NN Fold: ", end =" ")
            
            for i in range(fold):
                print (str(i+1), end =" "),
                #create training set
                partial_train_x = np.concatenate([train_x[:i * win],
                                                  train_x[(i + 1) * win:]], axis=0)
                partial_train_y = np.concatenate([train_y[:i * win],
                                                  train_y[(i + 1) * win:]], axis=0)
                #scale data here
                partial_train_x_scaled = (partial_train_x - 
                                          partial_train_x.mean(0))/partial_train_x.std(0)

                #fit training set
                self.fit(partial_train_x_scaled, partial_train_y)
                
                #create validation set
                val_x = train_x[i * win: (i + 1) * win]
                val_y = train_y[i * win: (i + 1) * win]
                #scale validation data here
                val_x_scaled = (val_x - partial_train_x.mean(0))/partial_train_x.std(0)
                
                #predict on validation set
                pred_y = self.predict(val_x_scaled, k_NN[j])
                val_acc = val_acc + self.accuracy(pred_y, val_y)
                
            val_acc = val_acc/fold
            if k_NN[j] < 10:
                print("[VAL ACC for 0" + str(k_NN[j]) + "-NN: " + str(val_acc)+"]")
            else:
                print("[VAL ACC for " + str(k_NN[j]) + "-NN: " + str(val_acc)+"]")
            k_acc.append(val_acc)
        return k_acc
        
    def predict(self, test_data, k):
        knn_train_x = self.knn_train_x
        knn_train_y = self.knn_train_y
        pred_class = []
        
        for i in range(len(test_data)):
            min_dist = {}
            #calculate all distance wrt to test_data[i]
            for key, value in knn_train_x.items():
                #check if test_data and train_data shape matches
                assert (np.shape(test_data[i]) == np.shape(value)), "Shape doesn't match!"
                kdist = np.linalg.norm(test_data[i] - value, ord=self.dist)
                min_dist[key] = kdist
            #sort all distances ascending order    
            min_dist_sort = sorted(min_dist.items(), key=operator.itemgetter(1))
            
            #choose argmax here for 3-KNN
            nn_x = []
            for knn in range(k):
                nn_x.append(knn_train_y[min_dist_sort[knn][0]])

            pred_class.append(max(nn_x,key=nn_x.count))
        return pred_class
    
    def accuracy(self, pred, test_y):
        test_y = test_y.reshape(len(test_y),).tolist()
        acc = 0
        for i in range(len(pred)):
            if(pred[i] == test_y[i]):
                acc = acc + 1

        return acc/len(pred)
```

## Question 3(c.ii) : Linear regression class implements fit, predict and accuracy


```python
class LinearRegression(object):
    def __init__(self):
        self.train_x = None
        self.train_y = None
        self.mean = None
        self.std = None
        self.weights = None
        print ("Linear Regression Model Intiallized...")
        
    #returns mean, var of std train set
    def scale_data(self, train_x):
        self.mean = train_x.mean(0)
        self.std = train_x.std(0)
        return self.mean, self.std
    
    #implements ordinary least square solution
    def fit(self, train_x, train_y):
        self.train_x = train_x
        self.train_y = train_y
        if (self.weights is None):
            train_square = np.matmul(self.train_x.transpose(), self.train_x)
            inv_train_transpose = np.matmul(np.linalg.inv(train_square),
                                            self.train_x.transpose())
            self.weights = np.matmul(inv_train_transpose, self.train_y)
        return self.weights
    
    def predict(self, test_data):
        return np.matmul(self.weights.transpose(), test_data.transpose()).transpose()
    
    #calculates RSS error 
    def accuracy(self, pred, test_y):
        return np.sum(np.square(test_y - pred))
```
